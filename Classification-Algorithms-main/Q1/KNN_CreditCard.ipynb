```python
import time
training_filename = 'creditcard_training.csv'
f = open(training_filename,'r')

dataset = []
labels = []
count = 0
start = time.time()
for x in f:
    temp_ls = []
    temp_s = ''
    for ch in x:
        if (ch >= '0' and ch <= '9') or ch == '.' or ch == '-' or ch == 'e':
            temp_s += ch
#             print('temp_s:' + temp_s)
        elif ch == ',':
#             print('temp_s:' + temp_s)
            if temp_s != '':
                temp_ls.append(float(temp_s))
            temp_s = ''
        elif ord(ch) == 10:
            continue
        else:
            break
#         print(ch + ': '+str(ord(ch)))
    
    if(len(temp_ls) == 30):
        dataset.append(temp_ls[1:-1])
    else:
        continue
    
    if '"0"' in x:
        labels.append(0)
    else:
        labels.append(1)
    
    count += 1

end = time.time()
print('time:' + str(end-start))
#     print('count:' + str(count))
```
