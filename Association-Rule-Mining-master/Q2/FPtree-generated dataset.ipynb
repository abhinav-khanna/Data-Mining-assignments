```python
#Code to read from the dataset file and obtain frequent 1-itemsets

filename = 'generated_dataset.txt'
f = open(filename,'r')
mp = []
mp.append({None:0})

min_sup  = 10   #Set minimum support = 0.01

temp_mp = {}
for x in f:   
    word = ''
    ls = []
    for i in range(len(x)):
        if (x[i] == '\n' or i == len(x) or ord(x[i]) == 10):
            ls = []
            if(word != ''):
                ls.append(word)
            if tuple(ls) in temp_mp.keys() and tuple(ls) != ():
                temp_mp[tuple(ls)] += 1
            else:
                temp_mp[tuple(ls)] = 1    
            word = ''
        elif(x[i] == ' ' or x[i] == ''):
            ls = []
            if(word != ''):
                ls.append(word)
            if (tuple(ls) in temp_mp.keys()) and tuple(ls) != ():
                temp_mp[tuple(ls)] += 1
            else:
                temp_mp[tuple(ls)] = 1 
            word = ''
        else:
            word += x[i]   
            
#Code to remove all infrequent items
for k in list(temp_mp):
    if temp_mp[k] < min_sup:
        print('removed element:' + str(k))
        temp_mp.pop(k)
        
mp.append(temp_mp)    
```

```python
#class declaration for node of a tree

class Node:
    support = 0
    key = None
    children = []
    neighbor = None
    cond_support = 0
    
    def __init__(self, key):
        self.key = key
        self.support = 1
        self.children = []
        self.neighbor = None
        self.cond_support = 0
```

```python
#pointer map: list of pointers acting as heads of linked list of nodes containing same item

pointer_map = {}
tail_map = {}

for i in mp[1].keys():
    pointer_map[i[0]] = None
    tail_map[i[0]] = None
```

```python
#items in a itemset should be ordered in decreasing order of support

def create_tuple(item_a, item_b):
    lsa = list(item_a)
    lsa.append(item_b)
    lsa.sort(key = compare_freq, reverse = True)
    
    return tuple(lsa)
    
```

```python
#Given a conditional FP tree, this function recusively calculates conditional support for an item
#Returns modified conditional tree

def count_cond_sup(item, root):
    
    if root == None:
        return (None, 0)
    
    if root.key == item[0]:
        if len(item) == 1:
            root.cond_support = 0
            return (root, root.support)
        else:
            temp = root.cond_support
            root.cond_support = 0
            return (root, temp)
    
    total_sum = 0
    
    for child in root.children:
        (node, cond_support) = count_cond_sup(item, child)
        total_sum += cond_support
    
    root.cond_support = total_sum
    return (root, root.cond_support)
```

```python
#Function to create a carbon copy of a conditional FP tree, to restore the original condtional tree

def copy_tree(root):
    if root == None:
        return None
    
    new_node = Node(root.key)
    new_node.support = root.support
    new_node.cond_support = root.cond_support
    new_node.neighbor = root.neighbor
    
    for child in root.children:
        new_child = copy_tree(child)
        new_node.children.append(new_child)
    
    return new_node
    
    
```

```python
#Function to restore the modified conditional tree to original conditional tree

def restore_tree(original_tree, restored_tree):
    if original_tree == None:
        return None
    
    restored_tree.cond_support = original_tree.cond_support
    
    for i in range(len(original_tree.children)):
        restored_tree.children[i] = restore_tree(original_tree.children[i], restored_tree.children[i])
    
    return restored_tree
```

```python
#Function to generate supersets of a given itemset

def generate_superset(item, root,all_items):
    (new_root, cond_sup) = count_cond_sup(item, root)
    
    #Create a xerox of the the new tree. Will be used later
    xerox_tree = copy_tree(new_root)
    
    
    if cond_sup >= min_sup:
        #Since conditional support is more than minimum support, add itemset to list of
        #frequent itemsets

        if freq_itemset != None:
            if (item not in freq_itemset[len(item)]):
                freq_itemset[len(item)][item] = cond_sup

            elif (freq_itemset[len(item)][item] == 0):
                freq_itemset[len(item)][item] = cond_sup
    
    else:
        #IMPORTANT: If itemset is not frequent, we can return from here as there is no need to check any superset
        #of this itemset
        return
    
    #iteratively check all itemsets by adding an extra item
    count = 0
    for obj in all_items:
        if obj not in item:
            tup = create_tuple(item, obj)
            if len(tup) == len(all_items):
                continue
            if tup not in freq_itemset[len(tup)]:
                #Note: restore tree since it would have been modified by 
                #calling generate superset
                new_root = restore_tree(xerox_tree, new_root)
                generate_superset(tup, new_root,all_items)
```
