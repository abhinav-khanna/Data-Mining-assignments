```python
import random

items = {}

num_items = 100  #set the number of total items
num_trans = 1000  #set the total number of transactions

# For item i, choose i number of random transactions that it will be included in
# This will generate a vertical transaction

for i in range(1,num_items):
    ls = []
    for j in range(i):
        tid = random.randint(1,num_trans)
        ls.append(tid)
    
    items[i] = ls

print(items)
```

```python
#Convert Vertical transactions to horizontal

transactions = {}

for i in range(1,num_trans+1):
    transactions[i] = []

for item in items.keys():
    ls = items[item]
    for i in ls:
        transactions[i].append(item)

print(transactions)
```

```python
#Write the generated horizontal dataset to a file, to run other algorithms
f = open('generated_dataset.txt','a')
for t in transactions.keys():
    ls = transactions[t]
    for item in ls:
        f.write(str(item) + ' ')
    
    f.write('\n')
```
