```python
#Code to read from the dataset file and obtain frequent 1-itemsets

filename = 'T10I4D100K.dat'
f = open(filename,'r')
mp = []
mp.append({None:0})

min_sup  = 1000    #Set minimum support = 0.01

temp_mp = {}
for x in f:   
    word = ''
    ls = []
    for i in range(len(x)):
        if (x[i] == '\n' or i == len(x) or ord(x[i]) == 10):
            ls = []
            if(word != ''):
                ls.append(word)
            if tuple(ls) in temp_mp.keys() and tuple(ls) != ():
                temp_mp[tuple(ls)] += 1
            else:
                temp_mp[tuple(ls)] = 1    
            word = ''
        elif(x[i] == ' ' or x[i] == ''):
            ls = []
            if(word != ''):
                ls.append(word)
            if (tuple(ls) in temp_mp.keys()) and tuple(ls) != ():
                temp_mp[tuple(ls)] += 1
            else:
                temp_mp[tuple(ls)] = 1 
            word = ''
        else:
            word += x[i]   
            
#Code to remove all infrequent items
for k in list(temp_mp):
    if temp_mp[k] < min_sup:
        print('removed element:' + str(k))
        temp_mp.pop(k)
        
mp.append(temp_mp)    
```

```python
#Function to take union and intersection of 2 tuples

def common_transaction(itemset_a,itemset_b):
    common_items = tuple(set(itemset_a).intersection(itemset_b))
    ls = list(common_items)
    ls.sort()
    common_items = tuple(ls)
    return (common_items)
```

```python
def union(itemset_a, itemset_b):
    tup = tuple(set(itemset_a + itemset_b))
    ls = list(tup)
    ls.sort()
    return tuple(ls)
```

```python
import time

total_items = len(mp[1])

for k in range(2, total_items + 1):
    
    start = time.time()
    print('k:' + str(k))
    temp_mp = {}
    for item_a in tidset[k-1].keys():
        for item_b in tidset[k-1].keys():
            if item_a != item_b:
                #Take transactions corresponding to 2 itemsets. If the intersection gives number of transactions
                #which greater than minimum support, then include the union of those 2 itemsets as frequent itemsets
                
                new_transactions = common_transaction(tidset[k-1][item_a],tidset[k-1][item_b])
                
                if len(new_transactions) >= min_sup:
                    new_itemset = union(item_a,item_b)
                    if (len(new_itemset) == k):
                        temp_mp[new_itemset] = new_transactions
    
    if temp_mp != {} and (temp_mp not in tidset):
        tidset.append(temp_mp)
    end = time.time()
    print('time taken:' + str(end-start))
    if temp_mp == {}:
        break
```

```python
#For each item, replace the mapping from the set of transactions to the cardinality of transactions

freq_itemset = []
freq_itemset.append({0:None})

for itemset in range(1,len(tidset)):
    temp_mp = {}
    for item in tidset[itemset].keys():
        temp_mp[item] = len(tidset[itemset][item])
    
    freq_itemset.append(temp_mp)
```
